```python
from scipy import stats
r, p = stats.pearsonr(df.Glucose, df.Outcome)
print(round(r, 4))
print(round(p, 42))
```

```python
import matplotlib.pyplot as plt 
plt.scatter(df.SkinThickness, df.BMI) 
plt.title('Skin Thicknes vs BMI')
plt.xlabel('Skin thicknes')
plt.ylabel('BMI')
plt.show()
```

```python
import matplotlib.pyplot as plt 
plt.scatter(df.Insulin, df.Glucose) 
plt.title('Insulin vs Glucose')
plt.xlabel('Insulin')
plt.ylabel('Glucose')
plt.show()
```

```python
df_diabetes = df[df['Outcome'] == 1]
df_nodiabetes = df[df['Outcome'] == 0]

plt.scatter(df_diabetes.Age, df_diabetes.Glucose, label='Diabetes: positive')
plt.scatter(df_nodiabetes.Age, df_nodiabetes.Glucose, label='Diabetes: negative')
plt.xlabel('Age')
plt.ylabel('Glucose')
plt.show()
```
